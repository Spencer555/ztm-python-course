```python
from sklearn.datasets import load_iris
iris = load_iris()
# we have imported the data in this case the data is already cleaned
```

```python
# now we split data  into train and test sets using X for test Y for train 
```

```python
X = iris.data # input
Y = iris.target # target 
# f(x) = y
# meaning we want to create a fucntion that takes input x and output data y

feature_names = iris.feature_names 
target_names = iris.target_names
target_names
# once we have understood the data we can split it into training and test set
```

```python
# start to split our data 
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4)
# the test size of 0.4 means 40 percent

print(X_train.shape)# return a dimention of the array 90 cos and 4 rows
print(X_test.shape)# return a dimention of the array 60 cos and 4 rows
```

```python
# create a model 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# WE KNOW THERE ARE 3 TYPES OF IRIS FLOWERS SO WE GIVE 3
knn = KNeighborsClassifier(n_neighbors=3)
dcn = DecisionTreeClassifier()
# now we fit the data with the train dataset 

knn.fit(X_train, Y_train)
dcn.fit(X_train, Y_train)
```

```python
# check the output
y_predicts = knn.predict(X_test)
#  we predict with the x test data and compare with the actual answers
```

```python
from sklearn import metrics
# this function takes the y_test which is our correct answers and y_predicts which is our prediction
# we can now check the accuracy of our model 
print(metrics.accuracy_score(Y_test, y_predicts))
# this gives us 0.95 which means 95% accurate
```

```python
# now we improve if we give it more test data than train data the accuracy reduces but this good for real world 
# have more cols e.g more features the more a machine can look at the better
# another to check is the algorithim 
```

```python
# decision tree classifier
D_predicts = dcn.predict(X_test)
print(metrics.accuracy_score(Y_test, D_predicts))
```

```python
# our own sample data
# predicting a flower we have not seen b4
sample = [[3,5,4,2],[2,3,5,4]]
predictions = knn.predict(sample)
predict_species = [iris.target_names[p] for p in predictions]
print('predictions: ', predict_species)
```

```python
# models are really time consuming and heavy on resources so 
# u can use in your code to receive input and predict something e.g whether an image is apporirate for a profile pic
# the predict model can predict and see if its apporpirate or not  and the function knn.fit()takes a long time and its done
# on gpu or cloud so we dont want every user that submit an image to run this 
# we call this model persistance so we want to save this model or function to a file and used it for predictions 
# we dont want to run and train the model every time a user want to submit an image
from joblib import dump, load

# saving the model
dump(knn, 'mlbrain.joblib')
# this stores the trained model in a file 
# so anytime we make an improvement we just dump it to the file
```

```python
# so in future instead of retraining the model we just do 

model = load('mlbrain.joblib')

# predictions = knn.predict(sample) instead of knn we now use model
model.predict(X_test) 
```
